```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt

# Assign the dataframe to this variable.
bmi_life_data = data = pd.read_csv("bmi_life_expectancy.csv")
bmi_life_data.plot(figsize=(15,5))
bmi_life_data.hist()
print(bmi_life_data.head(10))

# Check if there is missing data
print(bmi_life_data.isnull().values.any())

# Check for linear relationship
plt.scatter(bmi_life_data['Life expectancy'], bmi_life_data['BMI'])
# Make and fit the linear regression model
model = LinearRegression()
x_values = pd.DataFrame(bmi_life_data['Life expectancy'])
y_values = pd.DataFrame(bmi_life_data['BMI'])

bmi_life_model = model.fit(x_values, y_values)

# Make a prediction using the model
# TODO: Predict life expectancy for a BMI value of 21.07931
laos_life_exp = print(model.predict([[21.07931]]))

```
